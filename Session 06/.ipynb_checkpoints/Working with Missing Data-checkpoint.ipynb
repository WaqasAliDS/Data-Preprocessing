```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
happiness2015 = pd.read_csv("wh_2015.csv")
happiness2016 = pd.read_csv("wh_2016.csv")
happiness2017 = pd.read_csv("wh_2017.csv")
```

```python
happiness2017.columns = \
happiness2017.columns.str.replace("."," ").str.replace("\s+"," ").str.strip().str.upper()

happiness2017.columns
```

```python
happiness2016.columns = happiness2016.columns.str.replace("(","").str.replace(")","").str.strip().str.upper()

happiness2016.columns
```

```python
happiness2015.columns = happiness2015.columns.str.replace("(","").str.replace(")","").str.strip().str.upper()

happiness2015.columns
```

```python
combined = pd.concat([happiness2015, happiness2016, happiness2017], ignore_index = True)
```

```python
combined.isnull().sum()
```

```python
import seaborn as sns
```

```python
combined.set_index("YEAR")
```

```python
sns.heatmap(combined.set_index("YEAR").isnull(), cbar = False)

plt.show()
```

```python
regions = pd.read_csv("region.csv")
regions
```

```python
combined = pd.merge(combined, regions, on = "COUNTRY", how = "inner")
```

```python
combined[["REGION_x","REGION_y"]]
```

```python
combined.isnull().sum()
```

```python
combined = combined.drop("REGION_x", axis = 1)
```

```python
combined
```

```python
regions_columns = combined.pop("REGION_y")
```

```python
combined.insert(1, "REGIONS", regions_columns)
```

```python
combined
```

```python
combined[combined.duplicated(["COUNTRY","YEAR"])]
```

```python
"India" "INDIA"
```

```python
combined["COUNTRY"] = combined["COUNTRY"].str.upper()
```

```python
combined[combined.duplicated(["COUNTRY","YEAR"])]
```

```python
combined = combined.drop_duplicates(["COUNTRY","YEAR"])
```

```python
combined[combined.duplicated(["COUNTRY","YEAR"])]
```

```python
combined.isnull().sum()
```

```python
combined.notnull().sum().sort_values()
```

```python
combined = combined.dropna(thresh = 159, axis = 1)
```

```python
combined.notnull().sum().sort_values()
```

```python
combined.notnull().sum()
```

```python
sns.heatmap(combined.set_index("REGIONS").sort_values(["REGIONS", "HAPPINESS SCORE"]).isnull(), cbar = False)

plt.show()
```

```python
mean_family = combined["FAMILY"].mean()
mean_family
```

```python
combined["FAMILY"] = combined["FAMILY"].fillna(mean_family)
```

```python
combined["FAMILY"].mean()
```

```python
sns.heatmap(combined.set_index("REGIONS").sort_values(["REGIONS", "HAPPINESS SCORE"]).isnull(), cbar = False)

plt.show()
```
