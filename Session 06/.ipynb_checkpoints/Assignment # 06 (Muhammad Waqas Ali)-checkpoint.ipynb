# Dealing With Missing Data
The code performs various operations to clean and preprocess the World Happiness Report data from 2015, 2016, and 2017. It handles missing values, cleans column headers, merges with region data, drops unnecessary columns, removes duplicates

```python
def clean_headers(headers):
    cleaned_headers = headers.str.replace(".", " ").str.replace("\s+"," ").str.replace("(", "").str.replace(")", "").str.strip().str.upper()
    return cleaned_headers
```

Defines a function clean_headers to clean the column headers by replacing periods with spaces, removing extra spaces, parentheses, and converting them to uppercase.

```python
wh_2015.columns = clean_headers(wh_2015.columns)
wh_2016.columns = clean_headers(wh_2016.columns)
wh_2017.columns = clean_headers(wh_2017.columns)
```

Applies the clean_headers function to clean the column headers of each DataFrame.

```python
indexes_to_drop = drop_rows[~rows_to_keep].index
indexes_to_drop
```

Identifies the indexes of the rows in drop_rows that are to be dropped based on duplicated values in the "COUNTRY" and "YEAR" columns. The indexes are stored in the indexes_to_drop variable. In the context of dropping duplicates based on certain criteria, the ~ operator helps identify the rows that do not meet the criteria and need to be dropped.

```python
wh_combined = wh_combined.drop(indexes_to_drop)
```

```python
def fill_value(factors):
    wh_combined[factors] = wh_combined[factors].fillna(wh_combined[factors].mean())
    return 
```

This is a function definition named fill_value that takes a parameter factors. Inside the function, it fills the missing values in the columns specified by factors with the mean value of each column. The fillna() method is used to replace missing values with the mean value of the respective column.

```python
happiness_factors = wh_combined.columns[wh_combined.isnull().any()].tolist()
happiness_factors
```

This line creates a list named happiness_factors containing the column names that have at least one missing value. It uses the columns attribute of the wh_combined DataFrame to get the column names and the isnull().any() method to check if any missing values exist in each column.

```python
fill_value(happiness_factors)
```

This line calls the fill_value() function and passes the happiness_factors list as an argument. It fills the missing values in the columns specified by happiness_factors with the mean value of each column.

```python
sns.heatmap(wh_combined.set_index("REGION").sort_values(["REGION", "HAPPINESS SCORE"]).isnull(), cbar = False)
plt.show()
```

This code creates a heatmap using seaborn to visualize the missing values in the wh_combined DataFrame after filling missing values. The set_index("REGION") sets the "REGION" column as the index of the DataFrame for better visualization. The DataFrame is then sorted by "REGION" and "HAPPINESS SCORE" using the sort_values() method. The isnull() method is applied to identify missing values in the DataFrame, and the resulting Boolean DataFrame is passed to sns.heatmap() to create the heatmap. The cbar=False argument is used to hide the color bar. Finally, plt.show() displays the heatmap.
